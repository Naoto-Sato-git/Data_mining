演習問題1

```python
pr = 0.20
substandard = 0.80**10    # 10回受験して1度も合格しない確率
answer = (1 - substandard)*100
print("10回の受験で1度でも合格する確率は",round(answer,2),"%です")
```

演習問題3

```python
cards = 52
answer = 1
while(cards != 0):
  answer = answer*cards
  cards = cards - 1
print(f"トランプを52枚並べる並べ方は {answer:.2e} 通りです")
```

演習問題4

```python
p1=50
p2=60
p3=70
ave = (p1+p2+p3) / 3
dec = ((p1-ave)**2 + (p2-ave)**2 + (p3 - ave)**2) / 3
print("3人の体重の平均は",ave,"kgで、分散は",round(dec,2),"kgです")
```
