演習問題1

```python
import pandas as pd

game = pd.read_excel('game.xlsx')
game
```

```python
teens = game[game['年齢'] == '10代']
print("10代の学生のゲームプレイ時間の平均は",teens['プレイ時間'].mean(),"時間です")
```

```python
twenty = game[game['年齢'] == '20代']
print("20代の学生のゲームプレイ時間の平均は",twenty['プレイ時間'].mean(),"時間です")
```

```python
from scipy import stats
stats.ttest_ind(teens['プレイ時間'], twenty['プレイ時間'], equal_var = False)
```

pvalueが0.025より大きいので、10代と20代の学生のゲームプレイ時間には差がないと言える

演習問題2

```python
cross_table = pd.crosstab(game['年齢'], game['配信'])
cross_table
```

```python
from scipy import stats
alpha = 0.05  # 有意水準5%
df = 1        # クロス集計表が2行2列なので((2-1)*(2-1)=1)
stats.chi2.ppf(1-alpha, df)
```

```python
stats.chi2_contingency(cross_table, correction=True)
```

理論値が3.841 算出値が0.416

理論値が算出値と比較して大幅に大きいので、配信の有無と年齢は互いに独立した項目であることが分かる。(連関がない)
