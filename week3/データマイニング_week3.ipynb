```python
import requests
from io import StringIO
import pandas as pd

html = requests.get(
    url = "https://ja.wikipedia.org/wiki/%E5%85%83%E7%B4%A0",
    headers={"User-Agent":"Mozilla/5.0"}
).text

tables = pd.read_html(StringIO(html))

tables[1]
```

```python
tables[1].to_excel("elements.xlsx")
```
