演習問題1

```python
import numpy as np
from scipy import stats

alpha = 0.05 # 有意水準
sigma = 50   # 不偏標準偏差
n = 1000     # 標本数
df = n - 1   # 自由度
X = 500      # 帰無仮説
x = 450      # 標本平均
se = sigma / np.sqrt(n) # 標準誤差

interval = stats.t.interval(1 - alpha, df = df, loc = X, scale = se)
print("棄却域:" + str(interval))
if x < interval[0] or x > interval[1]:
  print("帰無仮説は棄却され、対立仮説が採択される")
  print("会社Aの平均年収は500万円ではない")
else:
  print("帰無仮説は受容され、対立仮説は採択されない")
  print("会社Aの平均年収は500万円である")
```

演習問題2

```python
alpha = 0.05 # 有意水準
sigma = 2    # 不偏標準偏差
n = 100      # 標本数
df = n - 1   # 自由度
X = 90       # 帰無仮説
x = 87       # 標本平均
se = sigma / np.sqrt(n) # 標準誤差

interval = stats.t.interval(1 - alpha, df = df, loc = X, scale = se)
print("棄却域:" + str(interval))
if x < interval[0] or x > interval[1]:
  print("帰無仮説は棄却され、対立仮説が採択される")
  print("この風邪薬を飲んで風邪が治る確率は90%ではない")
else:
  print("帰無仮説は受容され、対立仮説は採択されない")
  print("この風邪薬を飲んで風邪が治る確率は90%である")
```
