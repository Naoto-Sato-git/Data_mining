練習問題2

```python
father = 86
mother = 67
eldestDaughter = 73
eldestSun = 57
secondSun = 39
ave = (father + mother + eldestDaughter + eldestSun + secondSun) / 5
print("平均の貝の数は",ave,"個です。")
```

```python
import numpy as np
import statistics as stat

data = [86,67,73,57,39]
ave = np.mean(data)
max = np.max(data)
min = np.min(data)
mdn = stat.median(data)
# mdn = sorted(data)[int(len(data)/2)]

print("平均の貝の数は",ave,"個です。")
print("貝の数の最大値は",max,"個、中央値は",mdn,"個、最小値は",min,"個です。")
```

手計算で分散と不偏分散を計算

```python
sum = 0
for i in range (len(data)):
  sum += (data[i] - ave) ** 2
dis = sum / (len(data) - 1)
var = sum / (len(data))
print("標本分散は",var,"です。")
print("不偏分散は",dis,"です。")
```

コマンドで計算

```python
var = np.var(data)
dis = stat.variance(data)
print("標本分散は",var,"です。")
print("不偏分散は",dis,"です。")
```

地域Aと地域Bの家賃を比較するときに、平均だけで判断するのが難しい

一般的にはAの方が高いが、Aでは家賃の安い物件が多く選ばれ、Bでは高い物件が多く選ばれたりすると、平均は意味をなさない。
こういう時に、分散などが役に立つ

統計的検定

```python
from scipy import stats

alpha = 0.05 # 有意水準
sigma = 8 # 不偏標準偏差
n = 100 # 標本数
df = n - 1 # 自由度
X = 150 # 帰無仮説
x = 148 # 標本平均
se = sigma / np.sqrt(n) # 標準誤差

interval = stats.t.interval(1 - alpha, df = df, loc = X, scale = se)
print("棄却域:" + str(interval))
if x < interval[0] or x > interval[1]:
  print("帰無仮説は棄却され、対立仮説が採択される")
  print("ポテトチップスの平均内容量は150グラムではない")
else:
  print("帰無仮説は受容され、対立仮説は採択されない")
  print("ポテトチップスの平均内容量は150グラムである")
```
