AtCorder 第1問

2つの正整数a,bの積が偶数か奇数かを判別するプログラムを作成せよ

```python
#a = int(input("a:"));
#b = int(input("b:"));
a, b = map(int,input().split())
vol = a * b;
if(vol % 2 == 1):
  print("Odd")
else:
  print("Eve")
```

第2問

番号を持つマスが3つあり、1番のマスにのみビー玉を置ける。マスの番号からビー玉が何個置けるかを計算するプログラムを作成せよ。

```python
s = input("3桁の数字：")
c=0;
for i in s:
  if i == "1":
    c += 1
  else:
    continue
print("置かれるビー玉の個数は",c,"個");
```
