```python
def sortedlist(size):
  lst=[]
  for i in range(size):
    lst.append(i)
  return lst
```

演習問題1

```python
my_list=sortedlist(10)
even = []
odd = []
for i in my_list:
  if i%2==0:
    even.append(i)
  else:
    odd.append(i)
print("偶数：",even)
print("奇数：",odd)
```

演習問題2

```python
max=0
min=0
for i in my_list:
  if i > max:
    max = i
  elif i < min:
    min = i
  else:
    continue
print("最大値：",max)
print("最小値：",min)
```
