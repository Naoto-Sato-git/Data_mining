```python
health_data["年齢"]=health_data["年齢"].fillna(31)
health_data["血圧"]=health_data["血圧"].fillna(109)
health_data["肺活量"]=health_data["肺活量"].fillna(4037)
health_data["性別"]=health_data["性別"].fillna(1)
health_data["疾患の有無"]=health_data["疾患の有無"].fillna(0)
health_data["体重"]=health_data["体重"].fillna(64)
health_data
```
