```python
import pandas as pd
customer = pd.read_excel("../data_science_sample/4-2/顧客情報.xlsx")
customer
```

```python
item = pd.read_excel("../data_science_sample/4-2/商品情報.xlsx")
item
```

```python
transaction_4 = pd.read_excel("../data_science_sample/4-2/4月販売情報.xlsx")
transaction_4
```

```python
transaction_5 = pd.read_excel("../data_science_sample/4-2/5月販売情報.xlsx")
transaction_5
```

```python
transaction = pd.concat([transaction_4, transaction_5], ignore_index=True)
transaction
```

商品番号をもとにtransactionとitemを横に結合(JOIN)する

```python
join_data = pd.merge(transaction, item, on="商品番号")
join_data
```

```python
join_data = pd.merge(join_data, customer, on="顧客番号")
join_data = join_data.sort_values('販売日', ascending=True)
join_data = join_data.reset_index(drop=True)
join_data
```

データクレンジング

```python
uriage_data = pd.read_excel("../data_science_sample/4-3/クレンジング前.xlsx")
uriage_data
```

```python
uriage_data.loc[3, "曜日"] = "日"
uriage_data
```

```python
uriage_data["商品名"] = uriage_data["商品名"].str.replace("はさみ", "ハサミ")
uriage_data
```

```python
uriage_data["顧客の性別"].mode()
# modeはそのデータの指定された属性の最頻値(最もよく出てくる値)を指している。
#(今回は性別という属性において男よりも女性の方が多かった)
```

```python
uriage_data["顧客の性別"] = uriage_data["顧客の性別"].fillna("女性")
uriage_data
```

```python
uriage_data["顧客の年齢"].mean()
# meanはデータの中の平均値
```

```python
uriage_data["顧客の年齢"]=uriage_data["顧客の年齢"].fillna(16)
uriage_data
```
